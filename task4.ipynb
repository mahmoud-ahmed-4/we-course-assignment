```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Example data
data = {
    'Salary': ['high', 'medium', 'low', 'high', 'medium', 'high', 'low', 'high', 'medium', 'low']
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate the proportion of 'high' salaries
num_high_salaries = (df['Salary'] == 'high').sum()
total_salaries = len(df)
p_high = num_high_salaries / total_salaries

# Define parameters for the binomial distribution
n = total_salaries  # Number of trials (total number of salaries)
p = p_high          # Probability of success (having a high salary)

# Generate the binomial distribution
x = np.arange(0, n + 1)  # Possible number of successes
binom_pmf = binom.pmf(x, n, p)  # Probability mass function

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='skyblue')
plt.title('Binomial Distribution of High Salaries')
plt.xlabel('Number of High Salaries')
plt.ylabel('Probability')
plt.xticks(x)  # Show all x values
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Example data
data = {
    'Salary': ['high', 'medium', 'low', 'high', 'medium', 'high', 'low', 'high', 'medium', 'low']
}

# Create DataFrame
df = pd.DataFrame(data)

# Count the number of high salaries
num_high_salaries = (df['Salary'] == 'high').sum()

# Calculate the average rate (lambda) of high salaries
lambda_rate = num_high_salaries / len(df)

# Define the range of values for the Poisson distribution
x = np.arange(0, 5 * lambda_rate + 1)  # Extend the range for visualization

# Generate the Poisson distribution
poisson_pmf = poisson.pmf(x, lambda_rate)  # Probability mass function

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, color='skyblue')
plt.title('Poisson Distribution of High Salaries')
plt.xlabel('Number of High Salaries')
plt.ylabel('Probability')
plt.xticks(x)  # Show all x values
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Example data
data = {
    'Salary': ['high', 'medium', 'low', 'high', 'medium', 'high', 'low', 'high', 'medium', 'low']
}

# Create DataFrame
df = pd.DataFrame(data)

# Define the event of interest (e.g., having a 'high' salary)
event = 'high'

# Calculate the probability of having the 'high' salary
p = (df['Salary'] == event).mean()

# Define the range for the Bernoulli distribution
x = [0, 1]  # 0 for 'not high', 1 for 'high'

# Generate the Bernoulli distribution
bernoulli_pmf = bernoulli.pmf(x, p)  # Probability mass function

# Plot the distribution
plt.figure(figsize=(8, 5))
plt.bar(x, bernoulli_pmf, color='skyblue', tick_label=['Not High', 'High'])
plt.title('Bernoulli Distribution for Salary Data')
plt.xlabel('Salary Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns

# Load Titanic dataset
# You can use the Titanic dataset from seaborn's built-in datasets for this example
titanic_df = sns.load_dataset('titanic')

# Drop rows with missing values in 'survived' column for simplicity
titanic_df = titanic_df.dropna(subset=['survived'])

# Calculate the probability of survival (success)
p_survival = titanic_df['survived'].mean()

# Define the number of trials (n) and the probability of success (p)
n = len(titanic_df)  # Number of trials
p = p_survival  # Probability of success

# Define the range of values for the binomial distribution
x = np.arange(0, n + 1)  # Possible number of survivors

# Generate the binomial distribution
binom_pmf = binom.pmf(x, n, p)  # Probability mass function

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.bar(x, binom_pmf, color='skyblue')
plt.title('Binomial Distribution of the Number of Survivors on the Titanic')
plt.xlabel('Number of Survivors')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import seaborn as sns

# Load Titanic dataset
# You can use the Titanic dataset from seaborn's built-in datasets for this example
titanic_df = sns.load_dataset('titanic')

# Drop rows with missing values in 'survived' column for simplicity
titanic_df = titanic_df.dropna(subset=['survived'])

# Calculate the average rate (lambda) of survivors
lambda_rate = titanic_df['survived'].sum() / len(titanic_df) * len(titanic_df)  # Rate parameter ()

# Define the range of values for the Poisson distribution
x = np.arange(0, int(lambda_rate * 2))  # Extend the range to cover possible values

# Generate the Poisson distribution
poisson_pmf = poisson.pmf(x, lambda_rate)  # Probability mass function

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, color='skyblue')
plt.title('Poisson Distribution of the Number of Survivors on the Titanic')
plt.xlabel('Number of Survivors')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
